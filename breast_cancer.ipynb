```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}

# Re-initialize df from original data in case it was modified previously in the notebook.
# The 'data' variable is available from cell 32-h5JFAMY3E
df = pd.DataFrame(data)

# Identify the categorical column(s) to be encoded
categorical_columns = ['diagnosis']

for col in categorical_columns:
    if col in df.columns:
        le = LabelEncoder()
        # Fit and transform only the categorical column
        df[col] = le.fit_transform(df[col])
        # Store the fitted encoder for future use (e.g., inverse transform or new predictions for 'diagnosis')
        label_encoders[col] = le
```

```python
# split features and target
# 'id' is typically an identifier and not a feature for prediction.
# We drop 'diagnosis' as it's the target, and 'id' as it's not a feature.
x = df.drop(['diagnosis', 'id'], axis=1)
y = df['diagnosis'] #target
```
